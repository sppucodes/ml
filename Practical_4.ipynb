```python
x = 2
LR = 0.018
precision = 0.00001
prev_step_size = 1
max_iter = 10000
itr = 0
gd = []
```

```python
gf = lambda x : (x+3) ** 2
```

```python
while (precision < prev_step_size and itr < max_iter):
    prv = x
    x = x - LR* gf(prv)
    prev_step_size = abs(x - prv)
    itr += 1
    print("itr : ", itr, "value", x)
    gd.append(x)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(gd)
```
